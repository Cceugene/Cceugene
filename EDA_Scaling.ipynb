```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
df = pd.read_csv('/JupyterNotbook/data/sleep_file.csv')
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
how = pd.get_dummies(df['way'], prefix='how')
how.head()
```

```python
data = pd.concat([df, how], axis=1)
data = data.drop('way', axis=1)
data.head()
```
